```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
img = mpl.image.imread("tree.jpg")
```

```python
plt.imshow(img)
```

```python
img.shape
```

```python
flat_img = img.reshape(-1 , 3)
flat_img.shape
```

```python
kmean = KMeans(n_clusters=16 , n_init=5)
```

```python
kmean.fit(flat_img)
```

```python
compressed_img = kmean.cluster_centers_[kmean.labels_]
compressed_img = compressed_img.reshape(img.shape )
print("kmean_labels"  , kmean.labels_)
```

```python
plt.imshow(compressed_img/255)
```
